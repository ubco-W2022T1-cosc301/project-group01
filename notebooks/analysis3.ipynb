```python
# Task 1-3 Milestone 3

import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
data = data.drop(columns = ['gender','lunch','parental level of education'])

data['race/ethnicity']=data['race/ethnicity'].replace(['group A','group B','group C','group D','group E'],[1,2,3,4,5])
data
```

I dropped the "gender", "parental level of education", and "lunch" columns because they did not pertain to the topic of my research.

```python
sns.pairplot(data,hue='test preparation course')
```

## Fig 1.
This visual representation shows the relation between the different reading scores and whether or not a test preparation course was done. From this graph you can see the people who have completed test prep. courses tend to do better overall than people who did not. Similarly, there is a positive corelation between each pair of scores.

```python
sns.barplot(x='race/ethnicity',y='total score', data =data)
```

## Fig. 2
This graph shows the correlation between the different race/ethnicity groups and overall score. The graph shows that there is certain racial/ethnical groups tend to do better than others. Although there is not a significant difference, it is still worth noting.

```python
sns.displot(data,x="race/ethnicity",y="total score",hue = 'test preparation course',binwidth=(.5,2))
```

## Fig. 3
This graph shows the relation between different racial groups, the presence/lack of test prep courses. This graph shows how students from groups 4, 5, and 1 have larger amount of people who completed test prep courses and as such they were most prevalent in the higher scores. This supports the claim that there is disparity in the completion of test prep courses among racial/ethnical groups which results in better performance for said groups.
